```python
#10. Seperate Cabin column into num and char
df.head()
```

```python
df["Cabin"]
```

```python
character=[]
nums=[]
for i in df['Cabin']:
    s1=''
    s2=''
    if type(i)==str:
        for ch in i:
            if ch >= 'A' and ch <= 'Z' or ch >= 'a' and ch <= 'z':
                s1 += ch
            else:
                s2 += ch
        character.append(s1)
        nums.append(s2)
        
    else:
        character.append(i)
        nums.append(i)
```

```python
df['Cabin_char'] = character
df['Cabin_num'] = nums
```
